```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def my_dense(a_in, W, b):
  units=W.shape[1]
  a_out = sigmoid(np.dot(W.T, a_in)+b)
  return a_out
```

```python
print(my_dense([200, 13.9], W1_tmp, b1_tmp))
```

```python
# building two layers using my_dense subroutine
def my_sequential(x, W1, b1, W2, b2):
  a1=my_dense(x, W1, b1)
  a2=my_dense(a1, W2, b2)
  print(x.shape,W1.shape, a1.shape, W2.shape)
  print(f"a2: {a2.shape}")
  return a2
```

```python
W1_tmp = np.array( [[-8.93,  0.29, 12.9 ], [-0.1,  -7.32, 10.81]] )
b1_tmp = np.array( [-9.82, -9.28,  0.96] )
W2_tmp = np.array( [[-31.18], [-27.59], [-32.56]] )
b2_tmp = np.array( [15.41] )
```

```python
# writing a routine similar to model.predict() in tensorflow, it will take a matrix X with m examples
def my_predict(X, W1, b1, W2, b2):
  m=X.shape[0]
  p=np.zeros((m, 1))
  for i in range(m):
    print(my_sequential(X[i], W1, b1, W2, b2).shape)
    p[i, 0]=my_sequential(X[i], W1, b1, W2, b2)
  return p
```
